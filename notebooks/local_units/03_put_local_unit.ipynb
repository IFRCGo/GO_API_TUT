# PUT / POST Local Units — Emergency Response & Training and Education

**Environment:** Staging (`goadmin-stage.ifrc.org`)

This notebook contains two independent sections:

| Section | Method | Type | Description |
|---------|--------|------|-------------|
| A | POST | Emergency Response (type=3) | Creates a new Emergency Response unit |
| B | PUT  | Training and Education (type=5) | Fully replaces unit ID 116903 |

> WARNING: **This notebook targets STAGING. Production data will NOT be affected.**

**Local Unit Type codes (from API source):**
- 1 = Administrative
- 2 = Health Care
- 3 = Emergency Response
- 4 = Humanitarian Assistance Centres
- 5 = Training and Education
- 6 = Other

**Level (Coverage) codes:**
- 0 = National
- 1 = Regional
- 2 = District
- 3 = Local

---

## 1. Setup & Authentication

Run this cell first. It applies to both Section A and Section B.

```python
import requests
import json
import getpass
import datetime

# ============================================================
# ENVIRONMENT - Staging only (safe for testing)
# ============================================================
BASE_URL = "https://goadmin-stage.ifrc.org/api/v2"

TOKEN = getpass.getpass("Enter your IFRC GO STAGING API token: ")

HEADERS = {
    "Authorization": f"Token {TOKEN}",
    "Accept": "application/json",
    "Content-Type": "application/json"
}

TODAY = datetime.date.today().isoformat()  # e.g. '2026-03-04'

print("Authentication configured")
print(f"Environment : STAGING (goadmin-stage.ifrc.org)")
print(f"date_of_data: {TODAY}")
```

## Section A — POST: Create New Emergency Response Unit (type=3)

Sends a `POST` to `/api/v2/local-units/` to create a brand-new Emergency Response unit in Germany.

```python
# ============================================================
# SECTION A — Emergency Response unit data
# Fields visible in the UI form for this type:
#   date_of_data, sub-type, coverage (level), local unit name (En/Local),
#   phone, email, website, postcode, focal person (Local),
#   source (En/Local), address (En/Local), locality (En/Local),
#   latitude, longitude
# ============================================================

emergency_response_data = {
    # -- Required fields --
    "country": 72,                                         # 72 = Germany
    "type": 3,                                             # 3 = Emergency Response
    "level": 2,                                            # 2 = District
    "date_of_data": TODAY,                                 # Auto-set to today

    # -- Unit names --
    "local_branch_name": "DRK KV Freiburg Notfalleinheit",
    "english_branch_name": "DRK Freiburg Emergency Response Unit",

    # -- Location (required) --
    "location_json": {
        "lng": 7.8421,                                     # Longitude — Freiburg im Breisgau
        "lat": 47.9990                                     # Latitude
    },

    # -- Address --
    "address_loc": "Basler Strasse 2, 79100 Freiburg im Breisgau",
    "address_en": "Basler Street 2, 79100 Freiburg im Breisgau",
    "city_loc": "Freiburg im Breisgau",
    "city_en": "Freiburg",
    "postcode": "79100",

    # -- Contact --
    "phone": "+49 761 19222",
    "email": "notfall@drk-freiburg.de",
    "link": "https://www.drk-freiburg.de",

    # -- Focal person --
    "focal_person_loc": "Karl-Heinz Weber",
    "focal_person_en": "Karl-Heinz Weber",

    # -- Source --
    "source_loc": "DRK Kreisverband Freiburg e.V.",
    "source_en": "DRK District Association Freiburg",

    # -- Reason --
    "update_reason_overview": "New Emergency Response unit created via notebook on staging."
}

print("Section A — Emergency Response unit data:")
print(json.dumps(emergency_response_data, indent=2, ensure_ascii=False))
```

```python
# ============================================================
# SECTION A — Send POST request
# ============================================================

url_post = f"{BASE_URL}/local-units/"
print(f"Sending POST to: {url_post}")
print(f"   Environment: STAGING\n")

try:
    response_a = requests.post(url_post, headers=HEADERS, json=emergency_response_data)

    print(f"Response:")
    print(f"   Status Code  : {response_a.status_code}")
    print(f"   Content-Type : {response_a.headers.get('Content-Type', 'N/A')}")

    if response_a.status_code in [200, 201]:
        result_a = response_a.json()
        print(f"\nSUCCESS -- Emergency Response unit created on staging!")
        print(f"   New ID              : {result_a.get('id')}")
        print(f"   local_branch_name   : {result_a.get('local_branch_name')}")
        print(f"   english_branch_name : {result_a.get('english_branch_name')}")
        print(f"   Type                : {result_a.get('type')} (Emergency Response)")
        print(f"   date_of_data        : {result_a.get('date_of_data')}")
    else:
        print(f"\nFAILED -- Status {response_a.status_code}")
        print(f"   Response: {response_a.text[:500]}")
        try:
            error_a = response_a.json()
            if isinstance(error_a, dict):
                for key, value in error_a.items():
                    print(f"   {key}: {value}")
        except json.JSONDecodeError:
            pass

except requests.exceptions.RequestException as e:
    print(f"Request Error: {e}")
```

## Section B — PUT: Replace Training and Education Unit ID 116903 (type=5)

Sends a `PUT` to `/api/v2/local-units/116903/` to **fully replace** the existing Training and Education unit.

> WARNING: **PUT replaces the entire record. All fields must be present in the payload.**

```python
# ============================================================
# SECTION B — Fetch current data for unit 116903 (for reference)
# ============================================================

UNIT_ID_B = 116903
url_put = f"{BASE_URL}/local-units/{UNIT_ID_B}/"

print(f"Fetching current data for unit {UNIT_ID_B}...\n")

try:
    get_response_b = requests.get(url_put, headers=HEADERS)
    get_response_b.raise_for_status()
    current_b = get_response_b.json()

    print(f"Current unit data (before PUT):")
    print(f"   ID                  : {current_b.get('id')}")
    print(f"   local_branch_name   : {current_b.get('local_branch_name')}")
    print(f"   english_branch_name : {current_b.get('english_branch_name')}")
    print(f"   Type                : {current_b.get('type')} ({current_b.get('type_details', {}).get('name', 'N/A')})")
    print(f"   Status              : {current_b.get('status')}")
    print(f"   Country             : {current_b.get('country')}")
    print(f"   Level               : {current_b.get('level')}")
    print(f"   Address (local)     : {current_b.get('address_loc')}")
    print(f"   Address (en)        : {current_b.get('address_en')}")
    print(f"\nWARNING: This entire record will be REPLACED by the PUT request.")

except requests.exceptions.RequestException as e:
    print(f"ERROR: Could not fetch current data: {e}")
    current_b = None
```

```python
# ============================================================
# SECTION B — Training and Education unit data
# Fields visible in the UI form for this type:
#   date_of_data, sub-type, coverage (level), local unit name (En/Local),
#   phone, email, website, postcode, focal person (Local),
#   source (En/Local), address (En/Local), locality (En/Local),
#   latitude, longitude
# ============================================================

training_education_data = {
    # -- Required fields --
    "country": 72,                                         # 72 = Germany
    "type": 5,                                             # 5 = Training and Education
    "level": 1,                                            # 1 = Regional
    "date_of_data": TODAY,                                 # Auto-set to today

    # -- Unit names --
    "local_branch_name": "DRK Bildungszentrum Hannover",
    "english_branch_name": "DRK Training and Education Centre Hannover",

    # -- Location (required) --
    "location_json": {
        "lng": 9.7320,                                     # Longitude — Hannover
        "lat": 52.3759                                     # Latitude
    },

    # -- Address --
    "address_loc": "Podbielskistrasse 162, 30177 Hannover",
    "address_en": "Podbielski Street 162, 30177 Hannover",
    "city_loc": "Hannover",
    "city_en": "Hannover",
    "postcode": "30177",

    # -- Contact --
    "phone": "+49 511 928750",
    "email": "bildung@drk-hannover.de",
    "link": "https://www.drk-hannover.de",

    # -- Focal person --
    "focal_person_loc": "Sabine Hoffmann",
    "focal_person_en": "Sabine Hoffmann",

    # -- Source --
    "source_loc": "DRK Landesverband Niedersachsen e.V.",
    "source_en": "DRK State Association Lower Saxony",

    # -- Reason --
    "update_reason_overview": "Full replacement of Training and Education unit 116903 via notebook on staging."
}

print("Section B — Training and Education unit data (PUT payload):")
print(json.dumps(training_education_data, indent=2, ensure_ascii=False))
```

```python
# ============================================================
# SECTION B — Send PUT request
# ============================================================

print(f"Sending PUT to: {url_put}")
print(f"   Environment: STAGING\n")

try:
    response_b = requests.put(url_put, headers=HEADERS, json=training_education_data)

    print(f"Response:")
    print(f"   Status Code  : {response_b.status_code}")
    print(f"   Content-Type : {response_b.headers.get('Content-Type', 'N/A')}")

    if response_b.status_code in [200, 201, 204]:
        print(f"\nSUCCESS -- Unit {UNIT_ID_B} replaced on staging!")
        if response_b.status_code in [200, 201]:
            try:
                result_b = response_b.json()
                print(f"   ID                  : {result_b.get('id')}")
                print(f"   local_branch_name   : {result_b.get('local_branch_name')}")
                print(f"   english_branch_name : {result_b.get('english_branch_name')}")
                print(f"   Type                : {result_b.get('type')} (Training and Education)")
                print(f"   date_of_data        : {result_b.get('date_of_data')}")
            except json.JSONDecodeError:
                print("   (Could not parse response body)")
    else:
        print(f"\nFAILED -- Status {response_b.status_code}")
        print(f"   Response: {response_b.text[:500]}")
        try:
            error_b = response_b.json()
            if isinstance(error_b, dict):
                for key, value in error_b.items():
                    print(f"   {key}: {value}")
        except json.JSONDecodeError:
            pass

except requests.exceptions.RequestException as e:
    print(f"Request Error: {e}")
```
